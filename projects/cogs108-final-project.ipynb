## Background and Prior Work

Environmental threats like air pollution, carbon emissions, and climate change are increasingly recognized not just as ecological concerns, but as major public health issues. Around the world, people are experiencing the health effects of rising temperatures, worsening air quality, and more frequent extreme weather events. These environmental stressors have been linked to a range of health problems, especially respiratory diseases and certain types of cancer. As these issues become more widespread, understanding how environmental conditions impact health is critical—especially for informing policy decisions and protecting vulnerable populations<sup>1</sup>.

Previous research has highlighted strong connections between air pollution and serious health outcomes. For example, a review published in The Lancet describes air pollution as one of the leading global risk factors for disease and early death, especially in low- and middle-income countries where pollution levels are often highest<sup>1</sup>. Similarly, Kampa and Castanas (2008) explain how pollutants like particulate matter, nitrogen dioxide, and ozone contribute to health problems at the cellular level—causing inflammation, oxidative stress, and even DNA damage that can lead to cancer<sup>2</sup>. Additionally, projects like the "Cancer Level Prediction" on GitHub, which utilizes air pollution data to predict cancer levels based on environmental factors, demonstrate how machine learning can be used to link environmental threats with health outcomes<sup>3</sup>. These studies help make the case that environmental exposures are not just correlated with poor health outcomes, but can play a direct role in causing them.

Building on that work, our project looks at how environmental threats—including air pollution, rising carbon emissions, climate anomalies, and extreme weather events—relate to public health outcomes such as cancer rates and respiratory disease across different regions of the world. Our goal is to use data to find patterns in how these factors are connected, and to explore whether these insights can point to effective strategies for improving health outcomes, particularly in countries facing the greatest environmental challenges.

**Global Air Pollution Dataset**:
This dataset provides city-level air quality data across countries, with each observation containing numerical pollution metrics such as AQI Value, CO AQI Value, Ozone AQI Value, NO₂ AQI Value, and PM2.5 AQI Value (all floats). These variables serve as quantitative indicators of environmental pollution and will be used as independent variables in our analysis to assess their relationship with respiratory health outcomes. To prepare the dataset for analysis, we will remove any entries that do not correspond to recognized sovereign states by cross-referencing with a list of United Nations member countries. We will also drop any rows with missing values in key numerical columns. Since there are multiple observations per country, we will aggregate the data by computing the average value of each pollutant per country. Categorical pollution indicators (such as AQI Category or PM2.5 Category) will be excluded in favor of continuous values, which allow for more precise statistical analysis.

**Global Health Statistics Dataset**:
This dataset includes country-level health data on a variety of diseases, with key variables including Disease Name (string), Prevalence Rate, Incidence Rate, Mortality Rate, and Recovery Rate (all floats), as well as contextual factors such as Healthcare Access, Per Capita Income, and Urbanization Rate. These variables will serve multiple roles in our analysis: the disease-related rates will be used as dependent variables to measure the burden of respiratory illness, while healthcare access and socioeconomic indicators will function as control variables to account for differences in development and infrastructure across countries. The 'Disease Category' variable will be ignored due to inconsistencies and inaccuracies. Instead, to identify respiratory diseases—a central focus of our research—we will manually classify diseases using a curated list of known respiratory illnesses provided by the CDC (e.g., asthma, chronic obstructive pulmonary disease, pneumonia, bronchitis, etc.). Only diseases identified as respiratory in this list will be retained for analysis, allowing us to more accurately isolate the impact of air pollution on respiratory health.

**Climate Change Dataset**:
This dataset contains 9 columns capturing environmental and climate variables across various locations and dates. Key variables include 'Date' (string) which specifies the date of the observation, 'Location' (string) and 'Country' (string), categorical variables indicating where the data was recorded, 'Temperature' (float), 'CO2 Emissions' (float), 'Sea Level Rise' (float), 'Precipitation' (float), 'Humidity' (float), and 'Wind Speed' (float), numerical variables representing climate measurements. Since there are multiple observations per country, we will aggregate the data by computing the average value of each measure per country. 

**Combining the Datasets**:
We will merge the two datasets based on the country name field, ensuring that the final dataset includes only the countries present in both cleaned datasets. This merged dataset will allow us to investigate whether higher levels of air pollution are associated with greater prevalence, incidence, and mortality rates of respiratory diseases, while controlling for healthcare quality, economic status, and urbanization. Each variable has been selected to either contribute directly to testing our hypothesis (pollution and respiratory disease metrics) or to act as control variables to ensure our findings are not confounded by unrelated national differences.

## Global Air Pollution Dataset

```python
import pandas as pd
import numpy as np

# Read in global air pollution data 
air_pollution = pd.read_csv('global air pollution dataset.csv')

# Set 'Country' as the index and keep only numerical columns for analysis 
air_pollution_avg = air_pollution.set_index('Country').select_dtypes(include=np.number)

# Group the data by country and calculate the mean for each numeric column (in case there are multiple entries per country)
air_pollution_avg = air_pollution_avg.groupby('Country').mean()

# Clean up country names
air_pollution_avg.index = air_pollution_avg.index.str.strip().str.title()

# Define a mapping to standardize inconsistent or abbreviated country names to match a standard format
country_mapping = {
    "Russia": "Russian Federation",
    "South Korea": "Republic Of Korea",
    "Uk": "United Kingdom Of Great Britain And Northern Ireland",
    "Usa": "United States Of America",
    "UK": "United Kingdom Of Great Britain And Northern Ireland",  
    "USA": "United States Of America"  
}

# Apply the standardized country name mapping to the DataFrame index
air_pollution_avg = air_pollution_avg.rename(index=country_mapping)

# Display the resulting DataFrame with cleaned and averaged air pollution data by country
air_pollution_avg
```

## Global Health Statistics Dataset

```python
# Load the global health statistics dataset
global_health = pd.read_csv('Global Health Statistics.csv')

# Create a dictionary that maps specific diseases to broader disease categories
disease_categories = {
    "Malaria": "Vector-borne Infectious",
    "Ebola": "Infectious",
    "COVID-19": "Respiratory Infectious",
    "Parkinson's Disease": "Neurological",
    "Tuberculosis": "Respiratory Infectious",
    "Dengue": "Vector-borne Infectious",
    "Rabies": "Infectious (Neurological)",
    "Cholera": "Infectious (Waterborne)",
    "Leprosy": "Infectious (Chronic)",
    "Cancer": "Chronic (Non-communicable)",
    "Diabetes": "Chronic (Metabolic/Autoimmune)",
    "Measles": "Respiratory Infectious",
    "Zika": "Vector-borne Infectious (Neurological effects)",
    "Alzheimer's Disease": "Neurological",
    "Polio": "Infectious (Neurological)",
    "Hypertension": "Chronic (Cardiovascular)",
    "Asthma": "Chronic (Respiratory)",
    "HIV/AIDS": "Infectious (Immunodeficiency)",
    "Influenza": "Respiratory Infectious",
    "Hepatitis": "Infectious (Liver)"
}

# Use the mapping to create a new column called 'Disease Category'
global_health['Disease Category'] = global_health['Disease Name'].map(disease_categories)

# Drop the original 'Disease Name' column since we now have broader categories
global_health = global_health.drop(['Disease Name'], axis=1)

# Select only numeric columns to avoid aggregation errors
numeric_cols = global_health.select_dtypes(include='number').columns

# Combine numeric columns with the group-by columns
group_cols = ['Country', 'Disease Category']

# Group by country and disease category, then average only the numeric columns
global_health = global_health[group_cols + list(numeric_cols)].groupby(group_cols).mean().reset_index()

# Clean up country names 
global_health['Country'] = global_health['Country'].str.strip().str.title()

# Standardize certain country names using the same mapping dictionary used previously
global_health['Country'] = global_health['Country'].replace(country_mapping)

global_health
```

## Climate Change Dataset

```python
# Load the climate statistics dataset
climate_data = pd.read_csv('climate_change_data.csv')

# Clean up country names 
climate_data['Country'] = climate_data['Country'].str.strip().str.title()

# Standardize certain country names using the same mapping dictionary used previously
climate_data['Country'] = climate_data['Country'].replace(country_mapping)

# Group by Country and avergae columns
climate_data = climate_data.groupby('Country').mean(numeric_only=True)

climate_data
```

## Merged Health, Pollution, and Climate Data

```python
# Merge the three dataframes
merged = pd.merge(
    global_health, 
    air_pollution_avg, 
    right_index=True, 
    left_on='Country'
).merge(climate_data, left_on='Country', right_index=True)

# Drop the extra CO2 Emissions column
merged = merged.drop('CO2 Emissions', axis=1)
merged
```

### Correlation Analysis - Respiratory Infectious Diseases

To analyze the correlation between respiratory disease prevalence/incidence/mortality rates and the rest of our data, we will create scatterplots and a heatmap to look at the relationship between our variables of interest. We are choosing to look at all three disease indicators to see if trends between variables are consitent between the three measures. We will look for strong positive/negative trends to determine which parameters may be useful in our predictive model for respiratory disease. The correlation in the heatmap is measured by the r-value. 

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Define AQI columns and their labels
aqi_cols = [
    ("AQI Value", "Average AQI Value"),
    ("CO AQI Value", "Average CO AQI Value"),
    ("Ozone AQI Value", "Average Ozone AQI Value"),
    ("NO2 AQI Value", "Average NO2 AQI Value"),
    ("PM2.5 AQI Value", "Average PM2.5 AQI Value"),
]

# Create a 3x5 subplot grid
fig = make_subplots(
    rows=3, cols=5,
    subplot_titles=[
        "AQI vs Prevalence", "CO AQI vs Prevalence", "Ozone AQI vs Prevalence", "NO2 AQI vs Prevalence", "PM2.5 AQI vs Prevalence",
        "AQI vs Incidence", "CO AQI vs Incidence", "Ozone AQI vs Incidence", "NO2 AQI vs Incidence", "PM2.5 AQI vs Incidence",
        "AQI vs Mortality", "CO AQI vs Mortality", "Ozone AQI vs Mortality", "NO2 AQI vs Mortality", "PM2.5 AQI vs Mortality",
    ]
)

# Row 1: Prevalence Rate
for i, (col, label) in enumerate(aqi_cols, 1):
    fig.add_trace(
        go.Scatter(
            x=respiratory[col],
            y=respiratory['Prevalence Rate (%)'],
            mode='markers',
            marker=dict(size=6),
            name=label,
            showlegend=False
        ),
        row=1, col=i
    )
    fig.update_xaxes(title_text=label, row=1, col=i)
    fig.update_yaxes(title_text="Prevalence Rate (%)", row=1, col=i)

# Row 2: Incidence Rate
for i, (col, label) in enumerate(aqi_cols, 1):
    fig.add_trace(
        go.Scatter(
            x=respiratory[col],
            y=respiratory['Incidence Rate (%)'],
            mode='markers',
            marker=dict(size=6),
            name=label,
            showlegend=False
        ),
        row=2, col=i
    )
    fig.update_xaxes(title_text=label, row=2, col=i)
    fig.update_yaxes(title_text="Incidence Rate (%)", row=2, col=i)

# Row 3: Mortality Rate
for i, (col, label) in enumerate(aqi_cols, 1):
    fig.add_trace(
        go.Scatter(
            x=respiratory[col],
            y=respiratory['Mortality Rate (%)'],
            mode='markers',
            marker=dict(size=6),
            name=label,
            showlegend=False
        ),
        row=3, col=i
    )
    fig.update_xaxes(title_text=label, row=3, col=i)
    fig.update_yaxes(title_text="Mortality Rate (%)", row=3, col=i)

# Final layout adjustments
fig.update_layout(
    height=900, width=1500,
    title_text="Respiratory Disease Rates vs Air Quality Indexes by Country",
    margin=dict(t=80)
)

fig.show()
```

1. Prevalence vs AQI
- Across all AQI types, no strong relationship is observed with prevalence of respiratory infectious diseases.
- NO₂ and PM2.5 AQIs show a slightly upward spread, possibly indicating a weak positive association with prevalence, but the patterns are not consistent.
- CO AQI values are tightly clustered and show little to no correlation.

2. Incidence vs AQI
- CO, Ozone, and NO₂ AQIs show no clear trends with incidence.
- PM2.5 AQI may have a slight positive association with incidence, suggesting that countries with higher fine particulate pollution could have more new cases though this remains visually weak.
- General AQI and Ozone AQI show largely scattered data with no linear trend.

3. Mortality vs AQI
- Most pollutant AQIs, including CO, Ozone, and NO₂, show little to no correlation with mortality.
- Interestingly, there may be a slight negative trend between general AQI and mortality — suggesting countries with higher pollution levels have slightly lower respiratory mortality. However, this could be due to confounding variables (e.g., healthcare quality or reporting bias).
- PM2.5 AQI shows scattered results, with no strong linear relationship to mortality rates.

# Classifier to Predict Disease Category
In this project, we started with a dataset containing diseases classified into 14 specific categories. Our objective is to develop a machine learning classifier capable of predicting the disease category based on various environmental, socio-economic, and health-related predictors.

To improve the model’s generalizability and handle class imbalance more effectively, we grouped the original 14 categories into 3 broader, more general categories. These aggregated categories better capture the underlying disease patterns and allow the classifier to focus on distinguishing between major disease types rather than very granular subcategories.

By simplifying the classification task into these three main groups, the model can learn more robust patterns from the available data, which improves prediction performance and interpretability. This recategorization also aligns with epidemiological knowledge, where diseases often share similar risk factors and environmental influences within these broader groups.

## Methodology
### Data Preparation:
We selected a set of relevant predictors from the dataset, including environmental variables such as temperature, precipitation, and sea level rise, as well as socio-economic indicators like per capita income and healthcare access. We also engineered additional features to capture relationships such as the economic health index and fatality rate.

### Disease Grouping:
The original 14 disease categories were grouped into three main classes:
1. **Infectious & Vector-Borne Diseases**: 'chronic (cardiovascular)', 'chronic (respiratory)', 'respiratory infectious'
2. **Cancer & Chronic Diseases**: 'chronic (metabolic/autoimmune)', 'chronic (non-communicable)',
3. **Respiratory & Cardiovascular Diseases**: 'infectious', 'infectious (chronic)', 'infectious (immunodeficiency)', 'infectious (liver)', 'infectious (neurological)', 'infectious (waterborne)', 'vector-borne infectious', 'vector-borne infectious (neurological effects)'

### Model Training:
We used a Random Forest classifier due to its robustness and ability to handle complex, non-linear relationships between features. The data was split into training and testing sets with stratification to preserve class distributions. Categorical variables (e.g., country) were one-hot encoded, and numerical features were passed through without scaling as tree-based models do not require normalization.

### Evaluation Metrics:
The model was evaluated using precision, recall, F1-score, and accuracy, with special attention to the class imbalance through macro-averaged metrics.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Group disease categories to reduce classes
def categorize_disease_for_env_analysis(cat):
    cat = cat.lower()
    
    respiratory_cardiovascular = [
        'chronic (cardiovascular)',
        'chronic (respiratory)',
        'respiratory infectious'
    ]
    
    cancer_chronic = [
        'chronic (metabolic/autoimmune)',
        'chronic (non-communicable)',
    ]
    
    infectious_vector = [
        'infectious',
        'infectious (chronic)',
        'infectious (immunodeficiency)',
        'infectious (liver)',
        'infectious (neurological)',
        'infectious (waterborne)',
        'vector-borne infectious',
        'vector-borne infectious (neurological effects)'
    ]
    
    if any(sub in cat for sub in respiratory_cardiovascular):
        return 'Respiratory & Cardiovascular'
    elif any(sub in cat for sub in cancer_chronic):
        return 'Cancer & Chronic'
    elif any(sub in cat for sub in infectious_vector):
        return 'Infectious & Vector-Borne'
    else:
        return 'Other'

# Apply disease grouping
merged['Disease_Group'] = merged['Disease Category'].apply(categorize_disease_for_env_analysis)

# Feature selection
feature_cols = [
    'Country',
    'Prevalence Rate (%)',
    'Incidence Rate (%)',
    'Mortality Rate (%)',
    'Population Affected',
    'Average Treatment Cost (USD)',
    'Per Capita Income (USD)',
    'Urbanization Rate (%)',
    'Temperature',
    'Sea Level Rise',
    'Precipitation'
]

X = merged[feature_cols].copy()
y = merged['Disease_Group']

# Feature engineering: existing + new engineered features
X.loc[:, 'Economic_Health_Index'] = X['Per Capita Income (USD)'] / (X['Average Treatment Cost (USD)'] + 1)
X.loc[:, 'Climate_Stress'] = X['Temperature'] + X['Sea Level Rise'] + X['Precipitation']

# New engineered features:
X.loc[:, 'Fatality_Rate'] = X['Mortality Rate (%)'] / (X['Prevalence Rate (%)'] + 1e-6)
X.loc[:, 'Population_Urbanization'] = X['Population Affected'] / (X['Urbanization Rate (%)'] + 1e-6)
X.loc[:, 'Treatment_Income_Ratio'] = X['Average Treatment Cost (USD)'] / (X['Per Capita Income (USD)'] + 1e-6)

# Log-transform skewed features
for col in ['Population Affected', 'Average Treatment Cost (USD)', 'Per Capita Income (USD)']:
    X.loc[:, f'log_{col}'] = np.log1p(X[col])

# Prepare column lists for preprocessor
categorical_cols = ['Country']
numeric_cols = [col for col in X.columns if col not in categorical_cols]

# Encode target labels
le_target = LabelEncoder()
y_encoded = le_target.fit_transform(y)

# Define preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', numeric_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ])

# Define classifier pipeline
clf_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42, class_weight='balanced', n_estimators=100))
])

# Train/test split with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, stratify=y_encoded, test_size=0.2, random_state=42)

# Fit model
clf_pipeline.fit(X_train, y_train)

# Predict and evaluate
y_pred = clf_pipeline.predict(X_test)
print(classification_report(y_test, y_pred, target_names=le_target.classes_))
```

```python
import numpy as np
import pandas as pd

# Get the trained classifier from the pipeline
rf = clf_pipeline.named_steps['classifier']

# Get feature names after preprocessing
# numeric columns + one-hot encoded country columns
ohe = clf_pipeline.named_steps['preprocessor'].named_transformers_['cat']
cat_features = ohe.get_feature_names_out(categorical_cols)

all_feature_names = numeric_cols + list(cat_features)

# Get feature importances from the RandomForest
importances = rf.feature_importances_

# Create a DataFrame of features and their importance
feat_imp_df = pd.DataFrame({
    'feature': all_feature_names,
    'importance': importances
})

# Sort by importance descending
feat_imp_df = feat_imp_df.sort_values(by='importance', ascending=False).reset_index(drop=True)

print(feat_imp_df.head(20))  # top 20 features
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract trained RandomForestClassifier from pipeline
rf_model = clf_pipeline.named_steps['classifier']

# Get the preprocessor step to retrieve feature names
preprocessor = clf_pipeline.named_steps['preprocessor']

# Get numeric feature names (passed through)
numeric_features = numeric_cols

# Get categorical feature names after one-hot encoding
ohe = preprocessor.named_transformers_['cat']
cat_features = ohe.get_feature_names_out(categorical_cols)

# Combine all feature names
all_features = np.concatenate([numeric_features, cat_features])

# Get feature importances
importances = rf_model.feature_importances_

# Sort features by importance
indices = np.argsort(importances)[::-1]

# Plot top 18 features
top_n = 18
plt.figure(figsize=(12, 8))
plt.title("Top 18 Feature Importances from Random Forest")
plt.barh(range(top_n), importances[indices][:top_n][::-1], align='center', color='teal')
plt.yticks(range(top_n), all_features[indices][:top_n][::-1])
plt.xlabel("Feature Importance")
plt.tight_layout()
plt.show()
```

# Ethics & Privacy

Several ethical concerns are raised while analyzing global global health outcomes in relation to environmental threats, specifically regarding data privacy, consent, and the potential for biased representations within the datasets.

Firstly, health-related data, especially concerning illnesses such as cancer and respiratory diseases, is inherently sensitive. Although aggregated public health data is often anonymized, care must be taken to ensure that no personally identifiable information is used or inferred. Any datasets used must be ethically sourced, with clear documentation that consent and privacy standards were upheld during data collection.

Secondly, there is a substantial risk of representation bias in both environmental and health datasets. Wealthier countries typically have more robust healthcare infrastructures and reporting systems, resulting in more complete and accurate data on disease incidence and environmental monitoring. In contrast, lower-income countries may lack the resources for systematic data collection, leading to underreporting of both pollution exposure and health outcomes. This can introduce misleading patterns in the data, such as the appearance of a weaker correlation between environmental threats and health issues in underrepresented regions—not because the effect is absent, but because it is poorly measured.

Pollutants themselves may also be underreported in regions without comprehensive environmental monitoring networks. This can skew the analysis and potentially reinforce global health inequities if such data gaps are not acknowledged and addressed.
To mitigate these issues, part of our analytical approach will include cross-checking health and environmental data with country-specific sources where available, and incorporating economic indicators such as GDP per capita to control for systemic differences in health infrastructure and reporting capacity. Finally, we will be transparent about data limitations and confounding factors.
